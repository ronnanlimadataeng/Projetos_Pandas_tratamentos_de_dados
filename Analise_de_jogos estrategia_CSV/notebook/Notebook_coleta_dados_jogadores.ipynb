## Tratando ausencias nos dados

```python
dataframe.loc[dataframe['num_vitorias'].isnull()]
```

```python
# Realizando o calculo para mudar as variaveis que estão ausentes
vitorias = (dataframe.total_partidas - dataframe.num_derrotas)
derrotas = (dataframe.total_partidas - dataframe.num_vitorias)      
totalpartidas = (dataframe.num_vitorias + dataframe.num_derrotas)
```

notamos que era possivel preencher as linhas com dados ausentes fazendo calculos da diferença ou soma delas

```python
dataframe.num_vitorias.fillna(round(vitorias), inplace=True)
```

```python
dataframe.loc[dataframe['num_vitorias'].isnull()]
```

os dados foram finalmente substituidos e tratados para a coluna num_vitorias

```python
dataframe.isna().sum()
```

```python
dataframe.loc[dataframe['num_derrotas'].isnull()]
```

```python
dataframe.num_derrotas.fillna(round(derrotas), inplace=True)
```

```python
dataframe.loc[dataframe['num_derrotas'].isnull()]
```

os dados foram finalmente substituidos e tratados para a coluna num_derrotas

```python
dataframe.loc[dataframe['total_partidas'].isnull()]
```

# Querys para insesir os dados tratados nas tabelas(tb_jogo e tb_pais)

```python
for jogo_list_linha in jogo_list:
    try:
        query = f"""insert into tb_jogo (nome_jogo) 
                     values ('{jogo_list_linha}')"""
        cursor.execute(query)
        print(f"Registro inserido com sucesso: {jogo_list_linha}")
    except Exception as e:
        print(f"Não foi possível inserir o regristro {jogo_list_linha}. O erro foi encontrado foi: {e}")  
```

```python
for pais_list_linha in pais_list:
    try:
        query = f"""insert into tb_pais (nome_pais) 
                     values ('{pais_list_linha}')"""
        cursor.execute(query)
        print(f"Registro inserido com sucesso: {pais_list_linha}")
    except Exception as e:
        print(f"Não foi possível inserir o regristro {pais_list_linha}. O erro foi encontrado foi: {e}")  
```

### Recuperando os dados, verificando se estão retornando preenchidos do MYSQL

```python
df_pais_sql = pd.read_sql('SELECT * FROM tb_pais', con=connection)
df_pais_sql
```

```python
df_jogo_sql = pd.read_sql('SELECT * FROM tb_jogo', con=connection)
df_jogo_sql
```

### Vamos fazer um merge para unir as tabelas pais e jogo

```python
left = pd.DataFrame(df_pais_sql)
right = pd.DataFrame(df_jogadores_sem_idade)

tb_jogadores_com_pais = pd.merge(left, right, on="nome_pais")

tb_jogadores_com_pais
```

```python
left = pd.DataFrame(df_jogo_sql)
right = pd.DataFrame(tb_jogadores_com_pais)

tb_jogadores_com_jogo = pd.merge(left, right, on="nome_jogo")

tb_jogadores_com_jogo
```

```python
tb_jogadores_final = tb_jogadores_com_jogo.drop(['nome_jogo','nome_pais'], axis=1)
tb_jogadores_final = tb_jogadores_final[['nome_jogador', 'genero', 'data_nascimento',
                                         'num_vitorias','num_derrotas','total_partidas','cod_pais','cod_jogo']]
tb_jogadores_final
```

Para inserir a tabela final ao qual criamos depois de unir as demais tabelas, precisamos ordenar os titulos da tabela

```python
# criando a list para inserção de todas as linhas
cols = "`,`".join([str(i) for i in tb_jogadores_final.columns.tolist()])

# inserindo em um dataframe 1 gravação por vez
for tb_jogadores_final_linha,row in tb_jogadores_final.iterrows():
    try:
        sql = "INSERT INTO `tb_jogador` (`" +cols + "`) VALUES (" + "%s,"*(len(row)-1) + "%s)"
        cursor.execute(sql, tuple(row))
        connection.commit()
        print(f"Registro inserido com sucesso: {tb_jogadores_final_linha}")
    except Exception as e:
        print(f"Não foi possível inserir o regristro {tb_jogadores_final_linha}. O erro foi encontrado foi: {e}")    
```
